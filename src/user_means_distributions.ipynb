```python
######################################################################
# Modelling Sleep Duration Using Gaussian Processes
# LE49: MiniProject
# Jan Ondras (jo356), Trinity College
# 2017/2018
######################################################################
# For each user calculate mean of particular measure => show distributions
######################################################################


import numpy as np
import time
import glob
import matplotlib.pyplot as plt

t = 'sleepduration' # type of data to extract
# t = 'bedin'
# t = 'bedout'
# t = 'steps'

types = {
    'sleepduration': '10',
    'bedin': '11',
    'bedout': '12',
    'steps': '1',
    'weight': '2',
    'bloodpressure': '4',
    'heartrate': '7'
}
means = []
st = time.time()
# For each user
for file_name in glob.glob('./../Dataset/' + t + '/*.npz'):
    UID = file_name.split('/')[-1][:-4]
#     data = np.load(file_name)['xy']
    means.append( np.mean( np.load(file_name)['xy'][:,1] ) )
    
print "Time taken: ", time.time()-st
```

```python
print t, "mean =", np.mean(means)
srt_means = sorted(means)

plt.figure(figsize=(13,6))
plt.bar(range(len(means)), srt_means)
plt.title(t)
plt.legend(['Mean = '+str(np.mean(means))])
plt.show()

plt.figure(figsize=(13,6))
plt.title(t)
plt.hist(srt_means, 1000, normed=1, color='b')
plt.legend(['Mean = '+str(np.mean(means))])
plt.show()
```
